```python
def agent_stay(actor, actors_env, val):
  row,column,actor_type = actor
  if actor_type == "Pac-man":
    temp = actors_env["Ghost"][1][row][column]
    actors_env["Ghost"][1][row][column] = val
  else:
    temp = actors_env["Pac-man"][1][row][column]
    actors_env["Pac-man"][1][row][column] = val
  return actors_env, temp
```

```python
def step(actor, actors_env, temp=0, policy_based=False, learning_rate=0.001, gamma=0.8):
    row, column, actor_type = actor

    future_state = []
    q_vals_future = []
    q_vals_current = actors_env[actor_type][0][row][column]

    reward = actors_env[actor_type][1][row][column]
    if actor_type == "Pac-man" and reward == 10:
        actors_env["Pac-man"][1][row][column] = 0

    # Actions: Up, Down, Left, Right
    action_count = 0
    if column > 0:
        future_state.append((row, column - 1, actor_type))
        q_vals_future.append(actors_env[actor_type][0][row][column - 1])
        action_count += 1
        if actor_type == "Pac-man" and actors_env[actor_type][1][row][column - 1] >= 0:
            actors_env["Ghost"][1][row][column - 1] = 50
    if column < 18:
        future_state.append((row, column + 1, actor_type))
        q_vals_future.append(actors_env[actor_type][0][row][column + 1])
        action_count += 1
        if actor_type == "Pac-man" and actors_env[actor_type][1][row][column + 1] >= 0:
            actors_env["Ghost"][1][row][column + 1] = 50
    if row > 0:
        future_state.append((row - 1, column, actor_type))
        q_vals_future.append(actors_env[actor_type][0][row - 1][column])
        action_count += 1
        if actor_type == "Pac-man" and actors_env[actor_type][1][row - 1][column] >= 0:
            actors_env["Ghost"][1][row - 1][column] = 50
    if row < 18:
        future_state.append((row + 1, column, actor_type))
        q_vals_future.append(actors_env[actor_type][0][row + 1][column])
        action_count += 1
        if actor_type == "Pac-man" and actors_env[actor_type][1][row + 1][column] >= 0:
            actors_env["Ghost"][1][row + 1][column] = 50

    # Update Q-value using Bellman equation
    q_vals_current += learning_rate * (reward + gamma * max(q_vals_future) - q_vals_current)
    actors_env[actor_type][0][row][column] = q_vals_current

    # Policy-based or random action
    if policy_based:
        future_state = future_state[np.random.choice(np.flatnonzero(q_vals_future == np.max(q_vals_future)))]
    else:
        action = random.randint(0, action_count - 1)
        future_state = future_state[action]

    # Update perception in the other agent's environment
    if actor_type == "Pac-man":
        actors_env["Ghost"][1][row][column] = temp
        temp = actors_env["Ghost"][1][future_state[0]][future_state[1]]
        actors_env["Ghost"][1][future_state[0]][future_state[1]] = 100
    else:
        actors_env["Pac-man"][1][row][column] = temp
        temp = actors_env["Pac-man"][1][future_state[0]][future_state[1]]
        actors_env["Pac-man"][1][future_state[0]][future_state[1]] = -100

    return future_state, reward, actors_env, temp
```

```python
def policy(actor, actor_env, temp=0, policy_steps=5):
    """
    Executes a limited look-ahead policy for an agent.

    Args:
        actor (tuple): Current state of the agent (row, column, type).
        actor_env (dict): The environment state.
        temp (int, optional): Temporary state value. Defaults to 0.
        policy_steps (int, optional): Number of steps to simulate. Defaults to 5.

    Returns:
        tuple:
            - policy_states (list): List of states visited during the policy execution.
            - new_state (tuple): Final state of the agent after executing the policy.
            - reward_from_states (list): Rewards obtained at each step.
            - actor_env (dict): Updated environment state.
            - temp (int): Updated temporary state value.
    """
    policy_states = []
    reward_from_states = []

    for _ in range(policy_steps):
        # Perform a step using policy-based decision-making
        new_state, reward, actor_env, temp = step(actor, actor_env, temp, policy_based=True)

        # Track the state and reward
        policy_states.append((new_state[0], new_state[1]))  # Correctly track both row and column
        reward_from_states.append(reward)

        # Update the actor to the new state for the next step
        actor = new_state

    return policy_states, new_state, reward_from_states, actor_env, temp
```

```python
score = 0
reward_count = np.sum(reward_pac_man == 10)
actors_env["Pac-man"][1] = reward_pac_man.copy()
actors_env["Ghost"][1] = reward_ghost.copy()
actors = { "pac-man": (1, 1, "Pac-man"), "ghost 1": (17, 16, "Ghost"), "ghost 2": (16, 17, "Ghost") }

actors_env, temp1 = agent_stay(actors["pac-man"], actors_env, 100)
actors_env, temp2 = agent_stay(actors["ghost 1"], actors_env, -100)
actors_env, temp3 = agent_stay(actors["ghost 2"], actors_env, -100)

actors_actions = { "pac-man": [], "ghost 1": [], "ghost 2": [] }

while True:
    # Pac-man policy
    actors_actions["pac-man"], actors["pac-man"], rewards, actors_env, temp1 = policy(actors["pac-man"], actors_env, temp1, 20)
    score += sum(rewards)
    reward_count -= sum([reward > 0 for reward in rewards])

    if reward_count <= 0:
        print("Game Won!")
        print("Final Score =", score)
        break

    # Ghost policies
    actors_actions["ghost 1"], actors["ghost 1"], _, actors_env, temp2 = policy(actors["ghost 1"], actors_env, temp2, 20)
    actors_actions["ghost 2"], actors["ghost 2"], _, actors_env, temp3 = policy(actors["ghost 2"], actors_env, temp3, 20)
    actors_env["Ghost"][1] = np.where( (actors_env["Ghost"][1]>-100) & (actors_env["Ghost"][1]<100), (actors_env["Ghost"][1]//2), actors_env["Ghost"][1])

    # Check for collisions
    common = [actors_actions["pac-man"][i] == actors_actions["ghost 1"][i] for i in range(len(actors_actions["pac-man"]))]
    common += [actors_actions["pac-man"][i] == actors_actions["ghost 2"][i] for i in range(len(actors_actions["pac-man"]))]

    if any(common):
        print("Game Over!")
        print("Final Score =", score)
        break

    # Step-by-step progress
    for i in range(len(actors_actions["pac-man"])):
        print(f"Step('Pac-man'): {actors_actions['pac-man'][i]} || Step('Ghost 1'): {actors_actions['ghost 1'][i]} || "
              f"Step('Ghost 2'): {actors_actions['ghost 2'][i]} || Score = {score} || Remaining Reward = {reward_count}")
```
